```python
from jupyterthemes import jtplot
jtplot.style(theme = 'monokai', context = 'notebook', ticks = True, grid = False)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/tjjaf/Desktop/ML Projects/cervical_cancer_risk_data.csv')
```

```python
data.info()
```

```python
data = data.replace('?', np.nan)
data
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns = ['STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'])

data
```

```python
data = data.apply(pd.to_numeric)

data.dtypes
```

```python
data = data.fillna(data.mean())
```

```python
data.isnull().sum()
```

```python
correlation = data.corr()

plt.figure(figsize = (30, 30))
sns.heatmap(correlation, annot = True)
plt.show()
```

```python
Y = data.iloc[:, 33]
X = data.iloc[:, :32]
```

```python
X = np.array(X).astype('float32')
Y = np.array(Y).astype('float32')
```

```python
from sklearn.preprocessing import StandardScaler

stand_scal = StandardScaler()

X = stand_scal.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
!pip install xgboost
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(learning_rate = 0.01, max_depth = 800, n_estimators = 1200)

model.fit(X_train, Y_train)
```

```python
train_score = model.score(X_train, Y_train)
train_score
```

```python
test_score = model.score(X_test, Y_test)
test_score
```

```python
predictions = model.predict (X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

class_rep = classification_report(predictions, Y_test)

print(class_rep)

con_max = confusion_matrix(predictions, Y_test)

sns.heatmap(con_max, annot = True)
```
